```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Lead Scoring.csv')
print(df.shape)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['Converted'].value_counts().plot(kind='bar', color=['#E24B4A','#1D9E75'], ax=ax)
ax.set_title('Lead Conversion Distribution')
ax.set_xticklabels(['Not Converted', 'Converted'], rotation=0)
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
print(f"Conversion Rate: {df['Converted'].mean()*100:.1f}%")
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
df.groupby('Converted')['Total Time Spent on Website'].plot(
    kind='hist', bins=30, alpha=0.6, ax=ax, legend=True
)
ax.set_title('Time Spent on Website by Conversion')
ax.set_xlabel('Total Time Spent on Website (seconds)')
ax.legend(['Not Converted', 'Converted'])
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
source_conv = df.groupby('Lead Source')['Converted'].mean().sort_values(ascending=False)
source_conv.plot(kind='bar', color='#378ADD', ax=ax)
ax.set_title('Conversion Rate by Lead Source')
ax.set_ylabel('Conversion Rate')
ax.set_xlabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
occ_conv = df.groupby('What is your current occupation')['Converted'].mean().sort_values(ascending=False)
occ_conv.plot(kind='bar', color='#1D9E75', ax=ax)
ax.set_title('Conversion Rate by Occupation')
ax.set_ylabel('Conversion Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
numeric_df = df[['TotalVisits', 'Total Time Spent on Website', 
                  'Page Views Per Visit', 'Converted']].dropna()
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', 
            cmap='coolwarm', ax=ax)
ax.set_title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
import pickle

with open('models.pkl', 'rb') as f:
    bundle = pickle.load(f)

fi = bundle['feature_importances']
top10 = fi[:10]
names, scores = zip(*top10)

fig, ax = plt.subplots(figsize=(10, 5))
ax.barh(names[::-1], scores[::-1], color='#7F77DD')
ax.set_title('Top 10 Feature Importances (Random Forest)')
ax.set_xlabel('Importance Score')
plt.tight_layout()
plt.show()
```
